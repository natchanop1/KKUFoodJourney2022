## Popularity Based Recommender System

```python
ratings.merge(books,on='ISBN')
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating' : 'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df =num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings','Publisher']]
popular_df 
```

## Collaborative Filtering Based Recommender System

```python
import pickle
#pickle.dump(popular_df,open('popular.pkl','wb'))
```

```python
avg_rating_df = avg_rating_df.rename(columns={"Book-Title" : "Book_Title"})
```

```python
books = books.drop_duplicates('Book-Title')
books = books.rename(columns={"Book-Title" : "Book_Title", "Book-Author": "Book_Author", "Image-URL-M" : "Img_url" })

books
```

```python
num_rating_df = num_rating_df.rename(columns={"Book-Title" : "Book_Title"})
```

```python
new = books.merge(avg_rating_df,on='Book_Title').drop_duplicates('Book_Title')[['Book_Title','Book_Author','Img_url','avg_ratings', 'Publisher']]
new = new.merge(num_rating_df,on='Book_Title').drop_duplicates('Book_Title')[['Book_Title','Book_Author','Img_url','num_ratings','avg_ratings', 'Publisher']]
new = new[new['num_ratings']>=10].sort_values('avg_ratings',ascending=False)
new = new.head(10000)
```

```python
new['tags'] = new['Book_Author'] + new['Publisher'] 
new['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags'].values.astype('U')).toarray()
vector 
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```
